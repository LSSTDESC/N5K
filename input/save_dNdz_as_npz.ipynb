```python
import numpy as np
```

```python
dNdz_s_file = './dNdz_srcs_nbins3_fiducial_cutbins.dat'
dNdz_c_file = './dNdz_clust_nbins6_fiducial_cutbins.dat'
dNdz_npz_file = './dNdzs_3src_6lens.npz'

dNdz_s = np.loadtxt(dNdz_s_file)
dNdz_c = np.loadtxt(dNdz_c_file)
```

```python
z_sh = dNdz_s[:,0]
dNdz_sh = dNdz_s[:,1:]
```

```python
z_cl = dNdz_c[:,0]
dNdz_cl = dNdz_c[:,1:]
```

```python
np.savez(dNdz_npz_file, z_sh=z_sh, dNdz_sh=dNdz_sh, z_cl=z_cl, dNdz_cl=dNdz_cl)
```

```python
dNdz_file = np.load(dNdz_npz_file)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure()
for i in range(0,3):
    plt.plot(dNdz_file['z_sh'], dNdz_file['dNdz_sh'][:,i])
plt.show()
```

```python
plt.figure()
for i in range(0,6):
    plt.plot(dNdz_file['z_cl'], dNdz_file['dNdz_cl'][:,i])
plt.show()
```
