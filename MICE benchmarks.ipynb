```python
ll = np.loadtxt ("test_n5k_ell.txt")
Cl = np.loadtxt ("test_n5k_Cl.txt")
ClLimb_gsl = np.loadtxt ("test_n5k_ClLimb_gsl.txt")
ClLimb_scipy = np.loadtxt ("test_n5k_ClLimb_scipy.txt")
ClLimb = np.loadtxt ("test_n5k_ClLimb.txt")
Clcpp = np.loadtxt ("Cl_case1.dat")
```

```python
fig, (ax1, ax2) = plt.subplots(figsize=(10, 10), nrows=2, gridspec_kw={'height_ratios': [3,1]}, sharex=True)
ax1.loglog (ll, Cl)
ax1.loglog (ll, Cl)
ax1.loglog (ll, ClLimb_gsl)
ax1.loglog (ll, ClLimb_scipy)

ax2.loglog (ll, abs(Cl/ClLimb_gsl-1.))
ax2.plot (ll, abs(Cl/ClLimb_scipy-1.))
```

```python
fig, ((ax1, ax2)) = plt.subplots(figsize=(10, 10),  nrows=2, gridspec_kw={'height_ratios': [3,1]}, sharex=True)
ax1.loglog (ll, Cl)
ax1.loglog (Clcpp [:,0], Clcpp [:,1])

cmax = interp1d (ll, Cl)
ax2.loglog (Clcpp [:,0], abs(cmax (Clcpp [:,0])/(Clcpp [:,1])-1))
plt.show ()
```
