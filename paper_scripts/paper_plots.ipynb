Runtime vs number of cores.

```python
# Load values
Ncores, FKEM_times, FKEM_errors = np.loadtxt('./FKEM_time_ncores.dat', unpack=True)

Ncores, matter_times, matter_errors = np.loadtxt('./matter_time_ncores.dat', unpack=True)

Ncores, levin_times, levin_errors = np.loadtxt('./Levin_times_ncores_precomp.dat', unpack=True)
```

```python
plt.figure()
plt.errorbar(Ncores, FKEM_times,yerr=FKEM_errors[-1], color='#0072A0',linestyle='', marker='o', label='FKEM')
plt.errorbar(Ncores, levin_times,yerr=levin_errors[-1], color='k', linestyle='', marker='o', label='Levin')
plt.errorbar(Ncores, matter_times,yerr=matter_errors[-1], color='#cb9d06', linestyle='', marker='o', label='matter')
plt.plot(Ncores, FKEM_times[-1]*np.ones(len(FKEM_times)), color='#0072A0', linestyle = '--')
plt.plot(Ncores, levin_times[-1]*np.ones(len(levin_times)), color='k', linestyle = '--')
plt.plot(Ncores, matter_times[-1]*np.ones(len(matter_times)), color='#cb9d06', linestyle = '--')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('$N_{\\rm cores}$',fontsize=14)
plt.ylabel('run time (s)',fontsize=14)
plt.ylim(0.1,60)
#plt.xlim(60,70)
plt.legend()
#plt.show()
plt.tight_layout()
plt.savefig('./times_ncores.pdf')
```

```python
# Nbins scaling, fixed number of shear bins
ncl, nsh, t_Lev_sh1, err_Lev_sh1 = np.loadtxt('./times_nbins_100runs/Levin_with_precomp/times_nb_sh1_Levin.txt', unpack=True)
ncl, nsh, t_fkem_sh1, err_fkem_sh1 = np.loadtxt('./times_nbins_100runs/times_nb_sh1_FKEM.txt', unpack=True)
ncl, nsh, t_mat_sh1, err_mat_sh1 = np.loadtxt('./times_nbins_100runs/times_nb_sh1_MATTER.txt', unpack=True)

ncl5_Lev, nsh, t_Lev_sh5, err_Lev_sh5 = np.loadtxt('./times_nbins_100runs/Levin_with_precomp/times_nb_sh5_Levin.txt', unpack=True)
ncl5_FKEM, nsh, t_fkem_sh5, err_fkem_sh5 = np.loadtxt('./times_nbins_100runs/times_nb_sh5_FKEM.txt', unpack=True)
ncl5_mat, nsh, t_mat_sh5, err_mat_sh5 = np.loadtxt('./times_nbins_100runs/times_nb_sh5_MATTER.txt', unpack=True)
```

```python
plt.figure()
plt.errorbar(ncl, t_fkem_sh1, yerr=err_fkem_sh1, color='#0072A0', marker='v', linestyle='',label = 'FKEM, 1 shear bin')
plt.errorbar(ncl, t_Lev_sh1, yerr=err_Lev_sh1, color='k', marker='v', linestyle='',label = 'Levin, 1 shear bin')
plt.errorbar(ncl, t_mat_sh1, yerr=err_mat_sh1, color='#cb9d06', marker='v', linestyle='',label = 'matter, 1 shear bin')
plt.errorbar(ncl5_FKEM, t_fkem_sh5, yerr=err_fkem_sh5, color='#0072A0', marker='o', linestyle='',label = 'FKEM, 5 shear bins')
plt.errorbar(ncl5_Lev, t_Lev_sh5, yerr=err_Lev_sh5, color='k', marker='o', linestyle='',label = 'Levin, 5 shear bins')
plt.errorbar(ncl5_mat, t_mat_sh5, yerr=err_mat_sh5, color='#cb9d06', marker='o', linestyle='',label = 'matter, 5 shear bins')
plt.yscale('log')
#plt.xscale('log')
plt.ylim(0.01, 9)
plt.legend(fontsize='small', ncol=2, loc='lower right')
plt.xlabel('Number of redshift bins (lenses)', fontsize=14)
plt.ylabel('run time (s)', fontsize=14)
#plt.show()
plt.tight_layout()
plt.savefig('./time_vs_clustbins.pdf')
```

We now have an updated version of matter which sorts out an internal issue. Look at the half / quarter bin test now with this and with the double resolution for FKEM. (All bins.)

```python
FKEM, FKEM_all = np.loadtxt('./binwidth_fkem.dat', unpack=True)

MATTER, MATTER_all = np.loadtxt('./binwidth_matter.dat', unpack=True)

Levin, Levin_all = np.loadtxt('./binwidth_levin.dat', unpack=True)
```

```python
fig, ax = plt.subplots()
plt.semilogy(range(0,3), np.asarray(FKEM)**2, 'o', color='#0072A0',linestyle='-', label='FKEM')
plt.semilogy(range(0,3), np.asarray(Levin)**2, 'o', color='k', linestyle='-', label='Levin')
plt.semilogy(range(0,3), np.asarray(MATTER)**2, 'o',color='#cb9d06', linestyle='-', label='matter')
plt.semilogy(range(0,3), np.asarray(FKEM_all)**2, 'o', color='#0072A0', linestyle=':', label='FKEM all $\ell$')
plt.semilogy(range(0,3), np.asarray(Levin_all)**2, 'o', color='k', linestyle=':', label='Levin all $\ell$')
plt.semilogy(range(0,3), np.asarray(MATTER_all)**2, 'o', color='#cb9d06',linestyle=':', label='matter all $\ell$')
plt.ylabel('$\Delta \chi^2$', fontsize=14)
plt.ylim(0.05,200000)
labels = [item.get_text() for item in ax.get_xticklabels()]
labels[1] = 'Full'
labels[5] = 'Half'
labels[9] = 'Quarter'
#ax.tick_params(axis='y', labelsize=14)
ax.set_xticklabels(labels, fontsize=14)
plt.tight_layout()
#plt.title('All bins (updated matter, double FKEM res)')
plt.legend(ncol=2, loc='upper left')
plt.savefig('./binwidth.pdf')
#plt.show()
```
